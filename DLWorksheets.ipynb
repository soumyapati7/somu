```python
import numpy as np
import matplotlib.pyplot as plt

# Define the tanh function
def tanh_function(x):
    return (1 - np.exp(-2 * x)) / (1 + np.exp(-2 * x))

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Calculate y values using the tanh function
y_values = tanh_function(x_values)

# Plot the tanh function in cyan with dots
plt.plot(x_values, y_values, color='cyan', marker='o', linestyle='-', markersize=4)

# Title and labels
plt.title('Plot of tanh(x)')
plt.xlabel('x')
plt.ylabel('tanh(x)')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def z_function(x, y):
    return 1 / (1 + np.exp(-x - y))

# Generate values for x and y
x_values = np.linspace(-5, 5, 100)
y_values = np.linspace(-5, 5, 100)

# Create a meshgrid for x and y
x, y = np.meshgrid(x_values, y_values)

# Calculate z values using the defined function
z = z_function(x, y)

# Create 3D surface plot
fig = plt.figure(figsize=(12, 6))

# Surface plot
ax_surface = fig.add_subplot(131, projection='3d')
ax_surface.plot_surface(x, y, z, cmap='viridis')
ax_surface.set_title('3D Surface Plot')

# Contour plot
ax_contour = fig.add_subplot(132)
contour = ax_contour.contour(x, y, z, cmap='viridis')
ax_contour.set_title('Contour Plot')
plt.colorbar(contour, ax=ax_contour, orientation='vertical')

# Wireframe plot
ax_wireframe = fig.add_subplot(133, projection='3d')
ax_wireframe.plot_wireframe(x, y, z, cmap='viridis')
ax_wireframe.set_title('Wireframe Plot')

# Show the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def objective_function(x, y, w, b):
    x = np.asarray(x)
    y = np.asarray(y)
    return np.mean((w * x + b - y)**2)

def gradient_descent(x, y, w, b, learning_rate):
    dw = np.mean(2 * x * (w * x + b - y))
    db = np.mean(2 * (w * x + b - y))
    w -= learning_rate * dw
    b -= learning_rate * db
    return w, b

def momentum_gradient_descent(x, y, w, b, learning_rate, gamma, v_w=0, v_b=0):
    dw = np.mean(2 * x * (w * x + b - y))
    db = np.mean(2 * (w * x + b - y))
    v_w = gamma * v_w + learning_rate * dw
    v_b = gamma * v_b + learning_rate * db
    w -= v_w
    b -= v_b
    return w, b, v_w, v_b

def nesterov_accelerated_gradient_descent(x, y, w, b, learning_rate, gamma, v_w=0, v_b=0):
    w_ahead = w - gamma * v_w
    b_ahead = b - gamma * v_b
    dw = np.mean(2 * x * (w_ahead * x + b_ahead - y))
    db = np.mean(2 * (w_ahead * x + b_ahead - y))
    v_w = gamma * v_w + learning_rate * dw
    v_b = gamma * v_b + learning_rate * db
    w -= v_w
    b -= v_b
    return w, b, v_w, v_b

def plot_surface_contour(x, y, w_values, b_values, objective_func):
    W, B = np.meshgrid(w_values, b_values)
    Z = np.zeros_like(W)

    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            Z[i, j] = objective_func(x, y, W[i, j], B[i, j])

    fig = plt.figure(figsize=(12, 6))

    # Surface Plot
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.plot_surface(W, B, Z, cmap='viridis', alpha=0.8)
    ax1.set_xlabel('Weight (w)')
    ax1.set_ylabel('Bias (b)')
    ax1.set_zlabel('Objective Function')
    ax1.set_title('Surface Plot')

    # Contour Plot
    ax2 = fig.add_subplot(122)
    contour = ax2.contour(W, B, Z, levels=20, cmap='viridis')
    ax2.clabel(contour, inline=True, fontsize=8)
    ax2.set_xlabel('Weight (w)')
    ax2.set_ylabel('Bias (b)')
    ax2.set_title('Contour Plot')

    plt.show()

def main():
    x = np.array([3,4.25,4.5,4.9,5.5,6.1])
    y = np.array([0.018,0.182,0.269,0.450,0.731,0.900])

    w_values = np.linspace(-5, 5, 100)
    b_values = np.linspace(-5, 5, 100)

    x_choice = int(input("Enter the choice for x (1 for Gradient Descent, 2 for Momentum-based, 3 for Nesterov): "))
    learning_rate = float(input("Enter the learning rate: "))
    w = float(input("Enter the initial weight: "))
    b = float(input("Enter the initial bias: "))

    if x_choice == 1:
        for _ in range(100):
            w, b = gradient_descent(x, y, w, b, learning_rate)
    elif x_choice == 2:
        gamma = float(input("Enter the momentum parameter (gamma): "))
        v_w, v_b = 0, 0
        for _ in range(100):
            w, b, v_w, v_b = momentum_gradient_descent(x, y, w, b, learning_rate, gamma, v_w, v_b)
    elif x_choice == 3:
        gamma = float(input("Enter the momentum parameter (gamma): "))
        v_w, v_b = 0, 0
        for _ in range(100):
            w, b, v_w, v_b = nesterov_accelerated_gradient_descent(x, y, w, b, learning_rate, gamma, v_w, v_b)
    else:
        print("Invalid choice. Please enter 1, 2, or 3.")
        return

    plot_surface_contour(x, y, w_values, b_values, objective_function)

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def objective_function(x, y, w, b):
    x = np.asarray(x)
    y = np.asarray(y)
    return np.mean((w * x + b - y)**2)

def gradient_descent(x, y, w, b, learning_rate):
    dw = np.mean(2 * x * (w * x + b - y))
    db = np.mean(2 * (w * x + b - y))
    w -= learning_rate * dw
    b -= learning_rate * db
    return w, b

def plot_surface_contour(x, y, w_values, b_values, objective_func):
    W, B = np.meshgrid(w_values, b_values)
    Z = np.zeros_like(W)

    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            Z[i, j] = objective_func(x, y, W[i, j], B[i, j])

    fig = plt.figure(figsize=(12, 6))

    # Surface Plot
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.plot_surface(W, B, Z, cmap='viridis', alpha=0.8)
    ax1.set_xlabel('Weight (w)')
    ax1.set_ylabel('Bias (b)')
    ax1.set_zlabel('Objective Function')
    ax1.set_title('Surface Plot')

    # Contour Plot
    ax2 = fig.add_subplot(122)
    contour = ax2.contour(W, B, Z, levels=20, cmap='viridis')
    ax2.clabel(contour, inline=True, fontsize=8)
    ax2.set_xlabel('Weight (w)')
    ax2.set_ylabel('Bias (b)')
    ax2.set_title('Contour Plot')

    plt.show()


def main():
    x = np.array([3,4.25,4.5,4.9,5.5,6.1])
    y = np.array([0.018,0.182,0.269,0.450,0.731,0.900])

    w_values = np.linspace(-5, 5, 100)
    b_values = np.linspace(-5, 5, 100)

    x_choice = int(input("Enter the choice for x (1 for Gradient Descent, 2 for Momentum-based, 3 for Nesterov): "))
    learning_rate = float(input("Enter the learning rate: "))
    w = float(input("Enter the initial weight: "))
    b = float(input("Enter the initial bias: "))

    if x_choice == 1:
        for iteration in range(1, 4):
            w, b = gradient_descent(x, y, w, b, learning_rate)
            plot_surface_contour(x, y, w_values, b_values, objective_function, iteration)
            if iteration < 3:
                cont = input("Press Enter to continue to the next iteration...")

    elif x_choice == 2:
        gamma = float(input("Enter the momentum parameter (gamma): "))
        v_w, v_b = 0, 0
        for _ in range(100):
            w, b, v_w, v_b = momentum_gradient_descent(x, y, w, b, learning_rate, gamma, v_w, v_b)
    elif x_choice == 3:
        gamma = float(input("Enter the momentum parameter (gamma): "))
        v_w, v_b = 0, 0
        for _ in range(100):
            w, b, v_w, v_b = nesterov_accelerated_gradient_descent(x, y, w, b, learning_rate, gamma, v_w, v_b)
    else:
        print("Invalid choice. Please enter 1, 2, or 3.")
        return

    plot_surface_contour(x, y, w_values, b_values, objective_function)


if __name__ == "__main__":
    main()
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D

def calculate_params(filters, input_shape):
    model = Sequential()

    # 1x1 convolution
    model.add(Conv2D(filters[0], (1, 1), padding='same', activation='relu', input_shape=input_shape))

    # 3x3 convolution
    model.add(Conv2D(filters[1], (3, 3), padding='same', activation='relu', input_shape=input_shape))

    # 5x5 convolution
    model.add(Conv2D(filters[2], (5, 5), padding='same', activation='relu', input_shape=input_shape))

    # Max pooling
    model.add(MaxPooling2D((3, 3), strides=(1, 1), padding='same', input_shape=input_shape))

    # Average pooling
    model.add(AveragePooling2D((3, 3), strides=(1, 1), padding='same', input_shape=input_shape))

    # Build the model to initialize the layers
    model.build((None, *input_shape))

    # Calculate parameters for each operation
    params_1x1 = model.layers[0].count_params()
    params_3x3 = model.layers[1].count_params()
    params_5x5 = model.layers[2].count_params()
    params_max_pool = model.layers[3].count_params()
    params_avg_pool = model.layers[4].count_params()

    total_params = params_1x1 + params_3x3 + params_5x5 + params_max_pool + params_avg_pool

    return params_1x1, params_3x3, params_5x5, params_max_pool, params_avg_pool, total_params

# Example: Inception layer 3(b) filters
inception_3b_filters = [192, 256, 384]

# Input shape (assuming RGB image)
input_shape = (224, 224, 3)

# Calculate parameters
params_1x1, params_3x3, params_5x5, params_max_pool, params_avg_pool, total_params = calculate_params(inception_3b_filters, input_shape)

# Print results
print(f"Parameters for 1x1 convolution: {params_1x1}")
print(f"Parameters for 3x3 convolution: {params_3x3}")
print(f"Parameters for 5x5 convolution: {params_5x5}")
print(f"Parameters for Max pooling: {params_max_pool}")
print(f"Parameters for Average pooling: {params_avg_pool}")
print(f"Total parameters in Inception layer 3(b): {total_params}")
```
